# Methods

For my project, I utilized data from The World Banks [Gender Data Portal](https://genderdata.worldbank.org) which presents gender statistics, primarily from the United Nations and UNESCO. Its aim is to enhance comprehension of gender data and enable analyses that can inform policy decisions.

The dataset relies upon census data from member countries, surveys from other organizations, and existing research done by the United Nations. To streamline and standardize their data collection methods, the World Bank works with international organizations like the United Nations to adhere to strict standards of measurement. They also provide assistance to member countries in terms of technical assistance and data collection in turn for the World Bank being able to use the information that is collected.

The dataset contains various thematic indicators, each having entries for a variety of countries across the world. Each country also has entries for different years, ranging from 1970 to 2022. The indicators are split into different categories, such as education, health, etc. For this project, I decided to focus on the topics of _education_, _employment_, and _entrepreneurship_.

```python
#import statements
import pandas as pd
import matplotlib as plt
```

```python
from zipfile import ZipFile

zip_file = ZipFile('datasets.zip')
entre_df = pd.read_csv(zip_file.open('Entrepreneurship.csv'))
employ_df = pd.read_csv(zip_file.open('Employment.csv'))
health_df = pd.read_csv(zip_file.open('Health.csv'))
educ_df = pd.read_csv(zip_file.open('Education.csv'))

entre_ind = entre_df['Indicator Name'].unique()
employ_ind = employ_df['Indicator Name'].unique()
health_ind = health_df['Indicator Name'].unique()
educ_ind = educ_df['Indicator Name'].unique()
```

### Exploring the different years and countries: 

```python
countries = entre_df['Country Name']
pd.concat([countries, employ_df['Country Name']])
pd.concat([countries, health_df['Country Name']])
pd.concat([countries, educ_df['Country Name']])
countries = countries.unique()
countries = pd.Series(countries)
print("The different countries included are:")
print(countries)
```

```python
years = entre_df['Year']
pd.concat([years, employ_df['Year']])
pd.concat([years, health_df['Year']])
pd.concat([years, educ_df['Year']])
years = years.unique()
years = pd.Series(years)
print("The different years included are:")

print(years)
```
